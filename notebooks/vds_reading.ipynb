# VDS Reader

```python
import xarray as xr
```

```python
vds_path = "/glb/hou/ua.sepco/data.nobackup/xdigi/screenplay/mars_ursa/seismics/vds/f1c.b2_578764_SHL22MUOBS_URSACGVRTLSM0RMO_4D15N.vds"
```

```python
vdsds = xr.open_dataset(vds_path, engine="vds")
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(20, 8))

# Use imshow instead of the default plot
# Flip the vertical dimension by setting origin='upper'
img = vdsds.interp(crossline=3512.7).Amplitude.T.plot.imshow(
    cmap="seismic", 
    robust=True,  # This helps with color scaling
    origin='upper',  # This flips the vertical axis to start at 0,
    vmin=-1000,
    vmax=1000
)

# Add more descriptive labels
plt.xlabel('Inline')
plt.ylabel('Sample (ms)')

# Adjust colorbar and layout
plt.tight_layout()
```
